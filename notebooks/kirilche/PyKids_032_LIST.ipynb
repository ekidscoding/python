# –°–ø–∏—Å–∫–∏

–°–ø–∏—Å–∫–∏ - —Ü–µ —Ç–∞–∫—ñ –∑–º—ñ–Ω–Ω—ñ, —â–æ –º–æ–∂—É—Ç—å **–º—ñ—Å—Ç–∏—Ç–∏**:
*   –Ω—ñ—á–æ–≥–æ (–ø–æ—Ä–æ–∂–Ω—ñ–π —Å–ø–∏—Å–æ–∫) `[]`
*   1 –µ–ª–µ–º–µ–Ω—Ç –±—É–¥—å-—è–∫–æ–≥–æ —Ç–∏–ø—É `[1024]`, `["Mykyta"]`, `[("Ukraine", "Kyiv")]`
*   2+ –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –±—É–¥—å-—è–∫–æ–≥–æ —Ç–∏–ø—É `[1024, 365]`, `["Mykyta", "–û—Ä–∏—Å—è", "üòÄ"]`, `[("Ukraine", "Kyiv", 482), ("India", "New Delhi", 1931), ("Georgia", "Tbilisi", 455)]`

`in` - —Ü–µ –æ–ø–µ—Ä–∞—Ç–æ—Ä, —â–æ –ø–µ—Ä–µ–≤—ñ—Ä—è—î —á–∏ –º—ñ—Å—Ç–∏—Ç—å—Å—è –¥–µ—è–∫–µ –∑–Ω–∞—á–µ–Ω–Ω—è —É —Å–ø–∏—Å–∫—É.

````python
teachers = []

print(teachers)

print("Kiril" in teachers)
````

# append()

Append (–¥–æ–¥–∞—Ç–∏) - —Ü–µ –æ–ø–µ—Ä–∞—Ü—ñ—è –¥–æ–¥–∞–≤–∞–Ω–Ω—è –µ–ª–µ–º–µ–Ω—Ç—É –¥–æ —Å–ø–∏—Å–∫—É.

````python
teachers.append("Kiril")

print(teachers)

print("Kiril" in teachers)
````

# extend()

Extend (—Ä–æ–∑—à–∏—Ä–∏—Ç–∏, –ø–æ–¥–æ–≤–∂–∏—Ç–∏) - —Ü–µ –æ–ø–µ—Ä–∞—Ü—ñ—è –¥–æ–¥–∞–≤–∞–Ω–Ω—è –í–°–Ü–• –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –æ–¥–Ω–æ–≥–æ —Å–ø–∏—Å–∫—É –¥–æ —ñ–Ω—à–æ–≥–æ —Å–ø–∏—Å–∫—É. –ê–±–æ —ñ–Ω—à–∏–º–∏ —Å–ª–æ–≤–∞–º–∏ extend - —Ü–µ **–∑–ª–∏—Ç—Ç—è** —Å–ø–∏—Å–∫—ñ–≤

````python
teachers_2024 = ["Kiril", "Daryna"]
teachers_2021 = ["Anton", "Ark"]
# —Å—Ç–≤–æ—Ä–∏–º–æ –ü–°–ï–í–î–û–ù–Ü–ú –ø–µ—Ä—à–æ–≥–æ —Å–ø–∏—Å–∫—É
teachers_2025 = teachers_2024

teachers_2025.extend(teachers_2021)
print(f"{teachers_2025=}")
````

üü†–£–í–ê–ì–ê –∫–æ–¥ –≤–≥–æ—Ä—ñ –∑–º—ñ–Ω–∏—Ç—å —Ç–∞–∫–æ–∂ `teachers_2024`, –±–æ `teachers_2025` - —Ü–µ –≤—Å—å–æ–≥–æ –ª–∏—à **–ø—Å–µ–≤–¥–æ–Ω—ñ–º** –¥–ª—è `teachers_2024`

````python
print(f"{teachers_2024=}")
````

# copy()

–©–æ–± –Ω–µ —Ç—Ä–∞–ø–ª—è–ª–æ—Å—è —Ç–∞–∫, —â–æ –º–∏ –Ω–µ–Ω–∞—Ä–æ–∫–æ–º –∑–º—ñ–Ω–∏–ª–∏ –≤–∏—Ö—ñ–¥–Ω–∏–π —Å–ø–∏—Å–æ–∫, –º–æ–∂–Ω–∞ –∑–º—ñ–Ω–∏—Ç–∏ –æ–¥–∏–Ω —Ä—è–¥–æ–∫

````python
teachers_2024 = ["Kiril", "Daryna"]
teachers_2021 = ["Anton", "Ark"]
# —Å—Ç–≤–æ—Ä–∏–º–æ –ö–û–ü–Ü–Æ –ø–µ—Ä—à–æ–≥–æ —Å–ø–∏—Å–∫—É
teachers_2025 = teachers_2024.copy()

teachers_2025.extend(teachers_2021)
print(f"{teachers_2025=}")
print(f"{teachers_2024=}")
````

# remove()

> –Ü–Ω–¥–µ–∫—Å - —Ü–µ –ø–æ—Ä—è–¥–∫–æ–≤–∏–π –Ω–æ–º–µ—Ä, –ø–æ—á–∏–Ω–∞—é—á–∏ –∑ 0



````python
del teachers_2025[1]
print(f"{teachers_2025=}")
````

# pop()

–û–ø–µ—Ä–∞—Ü—ñ—è `pop()` –¥–æ–∑–≤–æ–ª—è—î –æ–¥–Ω–æ—á–∞—Å–Ω–æ –≤–∏–¥–∞–ª–∏—Ç–∏ –µ–ª–µ–º–µ–Ω—Ç —Ç–∞ –ø–æ–≤–µ—Ä–Ω—É—Ç–∏ –∑–Ω–∞—á–µ–Ω–Ω—è —Ü—å–æ–≥–æ –µ–ª–µ–º–µ–Ω—Ç—É.

````python
—á–µ—Ä–≥–∞_–Ω–∞_–Ω–∞–≥–æ—Ä–æ–¥–∂–µ–Ω–Ω—è = ["–í–∞—Å–∏–ª—å", "–ì–∞–ª–∏–Ω–∞", "–°–≤—è—Ç–æ—Å–ª–∞–≤", "–ù–µ—Å—Ç–æ—Ä"]

while —á–µ—Ä–≥–∞_–Ω–∞_–Ω–∞–≥–æ—Ä–æ–¥–∂–µ–Ω–Ω—è != []:
  –ª—é–¥–∏–Ω–∫–∞ = —á–µ—Ä–≥–∞_–Ω–∞_–Ω–∞–≥–æ—Ä–æ–¥–∂–µ–Ω–Ω—è.pop()
  print("–õ–∏—Å—Ç–æ–º –ø–æ–¥—è–∫–∏ –Ω–∞–≥–æ—Ä–æ–¥–∂—É—î—Ç—å—Å—è", –ª—é–¥–∏–Ω–∫–∞)

print("–í—Å—ñ—Ö –Ω–∞–≥–æ—Ä–æ–¥–∏–ª–∏")
````

# index()

````python
teachers_2021 = ["Anton", "Ark"]

if "Ark" in teachers_2021:
  ark_index = teachers_2021.index("Ark")
  message = f'''
  –ê—Ä–∫–∞–¥—ñ–π –∑–Ω–∞—Ö–æ–¥–∏—Ç—å—Å—è —É —Å–ø–∏—Å–∫—É –≤—á–∏—Ç–µ–ª—ñ–≤
  {teachers_2021=}
  –ø—ñ–¥ –Ω–æ–º–µ—Ä–æ–º {ark_index}, –∞–±–æ
  —è–∫—â–æ —Ä–∞—Ö—É–≤–∞—Ç–∏ —è–∫ –∑–≤–∏—á–∞–π–Ω—ñ –ª—é–¥–∏,
  —Ç–æ –≤—ñ–Ω {ark_index+1}–π"
  '''
  print(message)
````

# reverse()

````python
—á–µ—Ä–≥–∞_–Ω–∞_–Ω–∞–≥–æ—Ä–æ–¥–∂–µ–Ω–Ω—è = ["–í–∞—Å–∏–ª—å", "–ì–∞–ª–∏–Ω–∞", "–°–≤—è—Ç–æ—Å–ª–∞–≤", "–ù–µ—Å—Ç–æ—Ä"]

—á–µ—Ä–≥–∞_–Ω–∞_–Ω–∞–≥–æ—Ä–æ–¥–∂–µ–Ω–Ω—è.reverse()

while —á–µ—Ä–≥–∞_–Ω–∞_–Ω–∞–≥–æ—Ä–æ–¥–∂–µ–Ω–Ω—è != []:
  –ª—é–¥–∏–Ω–∫–∞ = —á–µ—Ä–≥–∞_–Ω–∞_–Ω–∞–≥–æ—Ä–æ–¥–∂–µ–Ω–Ω—è.pop()
  print("–õ–∏—Å—Ç–æ–º –ø–æ–¥—è–∫–∏ –Ω–∞–≥–æ—Ä–æ–¥–∂—É—î—Ç—å—Å—è", –ª—é–¥–∏–Ω–∫–∞)

print("–í—Å—ñ—Ö –Ω–∞–≥–æ—Ä–æ–¥–∏–ª–∏")
````
