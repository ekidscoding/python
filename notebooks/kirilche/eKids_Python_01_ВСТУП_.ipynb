# Формат Jupyter Notebook

Особливістю формату, з яким працює Colab є те, що блоки коду чередуються з блоками тексту.

Текст є довільним та слугує навчальним та інформативним цілям.

Програмні блоки (Code Cell) є або повнофункціональною програмою (рідше), або незалежною частиною цілої програми, яку можна виконати **окремо**.

Colab запамʼятовує результат роботи виконаного блоку коду та може використовувати цей результат у всіх наступних блоках.

```python
a = 5
b = 10

print(a + b)  # надрукує 15
```

Наступні блоки можуть:


*   ніяк не використовувати створене раніше
*   використати будь-що зі створеного раніше
*   перезаписати попередні значення новими

```python
print("Гарна погода, чи не так?")  # ніяк не використовує попередній код

# Перезапише значення b

b = 11

print("Перше число =", a)  # використовує значення з попереднього блоку
print("Друге число =", b)  # має нове значення

print("Добуток чисел =", a * b)
```

# Змінні

Імена, що пишуться ліворуч від знака `=` називають змінними. Змінна - це свого роду мітка або наліпка з іменем, яку ви можете присвоїти якомусь обʼєкту чи явищу.

Знак `=` - присвоїти.
Знак `==` - дорівнює.

`=` - оператор присвоєння, що виконає наступне:
*   візьме все, що праворуч та обчислить значення виразу. Всі імена змінних праворуч будуть замінені на їх значення.
*   надасть новому значенню імʼя, що написане ліворуч знака `=`

Потренуймося створювати змінні

```python
ocinka = 12  # присвоює значення 12 (числового типу, int) змінній 'ocinka'

zalik = "зараховано"  # присвоює значення рядкового типу, str змінній 'zalik'

# спочатку обчислиться вираз праворуч. Це буде склейка (конкатенація) рядків-слів
message = "Ваша оцінка: " + str(ocinka) + ".\n" + "Ваш залік: " + zalik

print(message)  # друкує повідомлення, яке називає змінна message
```
