# Assignment (Присвоєння)

Оператор **присвоєння** `=` це не те саме, що в математиці знак *дорівнює*
Порівняння двох значень відбувається символом `==`, що складається з двох знаків =

```python
a: int = 123
b: float = 4.56
c: bool = True
d: str = "Let's GO!"
e: list = []  # створить порожній список.
# e: list = [1, 2, 3, 4, 5, 6]
# або якщо це послідовні числа
# e: list = list(range(1, 7))
f: tuple = 1, "dog", True, [1, 2, 3], 5.24
# або
# f: tuple = (1,) – створення одноелементного кортежу.
g: dict = {}  # створить порожній словник
h: set = set()  # створить порожню множину
i: frozenset = frozenset(h)  # статична, незмінна множина
j: None = None

print(a, b, c, d, e, f, g, h, i, j, sep='\n')
```

# анатомія присвоєння


*   a: int = 123

  *   де `a` - імʼя змінної
  *   `: int` - необовʼязкова анотація типу. Якщо її не вказати - нічого не зміниться. Але підказка типу допомагає читати код і розуміти його
  *   власне `=` - знак оператора **присвоїти**
  *   `123` - довільний вираз,

      що **обчислюється** в ціле число. Може бути `5 + 17`
  
      або `8**2 - 5 * 6 + 3 - 126 % 19 + 7 * (7 // 3)`



```python
my_int = 8**2 - 5 * 6 + 3 - 126 % 19 + 7 * (7 // 3)
print(f"My integer number value is {my_int}")
```

# анатомія присвоєння (продовження)

*   `i = frozenset(h)`
    *   `i` - імʼя змінної
    *   `frozenset()` - **виклик функції**

        тут результат виклику функції, те, що вона повертає буде названо іменем змінної, яку ми вказали ліворуч.

*   `my_int = int(input("Введіть ціле число: "))`
    *   `my_int` -
    *   `input(...)` - внутрішня функція. За правилами **композиції функцій** саме вона буде виконана першою.
    *   `int(...)` - виклик зовнішньої функції, яка аргументом має те, що повернула внутрішня функція

```python
# @title Композиція функцій

my_int = int(input("Введіть ціле число: "))
print(f"Наступне ціле число = {my_int + 1}")
print(f"Квадрат вашого цілого числа = {my_int ** 2}")
```

# Chain Assignment (Присвоєння ланцюжком)
Ми можемо присвоювати декілька змінних одночасно. Несе такий зміст: значення однієї та іншої змінної мають стати значенням виразу, що стоїть праворуч

```python
sum1 = sum2 = 0
print(f"{sum1=}, {sum2=}")
```

# Примітки


1.   Якщо в f-рядку в фігурних дужках вказати імʼя змінної та знак `=`, ми побачимо у виводі імʼя змінної та її значення. Також всередині фігурних дужок можливе вирівнювання для більш структурованого виводу типу таблиці.
2.   Присвоювання ланцюжком не є розповсюдженим та може давати **неочікуваний результат** коли працюємо зі списками (див. приклад)



```python
list1 = list2 = []
print(f"Після присвоєння ланцюжком: {list1=}, {list2=}")
print("Додаємо новий елемент 'A' до списку 'list1'")
list1.append('A')
print(f"{list1=}, {list2=} <--- ЧОМУ?")

print("\nРозгадка в тому, що ці два списки --\nпросто ПСЕВДОНІМИ одного й того ж обʼєкту,\nв чому можна пересвідчитись за допомогою функції id()\n")
print(f"{id(list1)=}, {id(list2)=}")
```

# Parallel Assignment (aka Tuple Unpacking)
Паралельне присвоєння, або розпакування кортежу. Це такий процес, коли **ліворуч** та **праворуч** від знака присвоєння `=` ми пишемо

> однакову кількість імен змінних та виразів відповідно (N імен змінних ліворуч та відповідно N виразів праворуч від `=`)

Це зручно у багатьох випадках та абсолютно **необхідно** для

*  роботи зі складними типами даних та
*  функціями, що повертають декілька значень.

```python
a, b = 10, 46
print(f"{a=}, {b=}")
a, b = b, a
print(f"{a=}, {b=}")
```

```python
# @title Частка та остача
dilene = input("Введіть ділене: ")
dilnyk = input("Введіть дільник: ")
chastka, ostacha = divmod(int(dilene), int(dilnyk))

print(f"{dilene} ділиться на {dilnyk} з часткою {chastka} та остачею {ostacha}")
print(f"{dilene} = {dilnyk} * {chastka} + {ostacha}")
```
