# String (рядок) та List (Список)
Рядок - це складний тип даних, який

*   є послідовність (`sequence`)
*   НЕзмінний (`immutable`)
*   складається з окремих елементів (`container` type)
*   кожен елемент обов'язково є одним символом (`character` aka "char")



Список - це складний тип даних, який
*   є послідовність (`sequence`)
*   змінний (`mutable`)
*   складається з окремих елементів (`container` type)

Змінний (`mutable`) - такий, що ми можемо змінити без перестворення.

Незмінний (`immutable`) - це такий тип даних, який не змінюється без перестворення.

Типовим представником незмінного типу даних є `str` (string).

list навпаки ми можемо змінювати на місці (`in-place`)

```python
my_str = 'abc'
print(my_str[0])

print(my_str[1] == 'b')

print(my_str[-1] == 'c')

# наступний рядок викличе помилку, тому що рядок - НЕзмінний (immutable) тип
my_str[0] = 'k'
```

```python
my_list = list(my_str)
print(my_list)

print(my_list[1] == 'b')

print(my_list[-1] == 'c')

# наступний рядок НЕ викличе помилку, тому що список - змінний (mutable) тип
my_list[0] = 'k'

print(my_list)
```

# Slice (зріз)

Slice - це частина списку або рядку. Щоб створити слайс ми вказуємо в квадратних дужках 2 значення через `:`

При цьому слайс включатиме початкове значення та не вбирає кінцеве.

```python
my_str = 'AaBbCcDdEeFf'

start = 0
stop = 5

print(f"Почнемо з '{my_str[start]}' включно")
print(f"Дійдемо до '{my_str[stop]}' але НЕ включно")
```

# Надрукувати символи рядку ВІД 'start' ДО 'stop'

```python
# @title класичний метод, за допомогою цикла
i = start
while i < stop:
  print(my_str[i], end='')
  i += 1
print()
```

```python
# @title те саме, але з slice
print(f"Скорочено та більш зручно {my_str[start:stop]=}")
```

# slice може ще так

якщо вкажемо додатковий третій параметр, то зможемо **міняти напрям** та **крокувати**

`start = None` значить присвоїти змінній `start` спеціальне значення `None` яке значить "Відсутність Значення"

```python
start = None
stop = None
step = -1

print(f"Скорочено та більш зручно {my_str[start:stop:step]=}")

print(f"Те саме, але без `None` взагалі {my_str[::step]=}")
```

```python
start = 0
stop = -1
step = 2

print(f"Скорочено та більш зручно {my_str[start:stop:step]=}")

print(f"Те саме, але без start=0 та stop=-1 взагалі {my_str[::step]=}")
```
