## Пишемо код

**Кодинг** (написання програмного коду) - це досить простий процес, якщо є **дизайн програми**. Якщо такого немає, то кодинг може перетворитись на тривалу серію проб та помилок.

Суть кодингу полягає в тому, щоб перетворити дизайн програми на цілком реальну програму, яка успішно **виконується і влаштовує** користувачів.

Зауважте, користувачами програми є не лише ті, хто її запускають, але й ваші колеги, які її можуть читати та дописувати, додаючи функціонал в майбутньому.

### Python Coding
Напишемо те, що ми описали вище коли дизайнили програму.

```python
# підключаємо генератор випадкових чисел
from random import randint

# базові налаштування
minimum, maximum = 1, 10
# вміти загадувати випадкове (random) число
random_number = randint(minimum, maximum)

print(f'Введіть число між {minimum} та {maximum}.')

# поки користувач не вгадає число
while True:
  # запитувати користувача "Введіть число"
  user_guess = int(input('Введіть число: '))

  # постійно порівнювати загадане випадкове число з тою здогадкою (guess),
  # яку ввів користувач
  if user_guess < random_number:
    # якщо здогадка менша за загадане - вивести "Замало"
    print('Замало')

  if user_guess > random_number:
    # якщо здогадка більша за загадане - вивести "Забагато"
    print('Забагато')

  if user_guess == random_number:
    # якщо здогадка вірна (дорівнює загаданому) - привітати користувача
    print('Ви вгадали число')
    # перестати запитувати новий ввід
    break
```

# Власне сама програма (Гра)

```python
from random import randint

minimum, maximum = 1, 10
random_number = randint(minimum, maximum)

print(f'Введіть число між {minimum} та {maximum}.')

while True:
  user_guess = int(input('Введіть число: '))

  if user_guess < random_number:
    print('Замало')

  if user_guess > random_number:
    print('Забагато')

  if user_guess == random_number:
    print('Ви вгадали число')
    break
```

# Розбір механіки гри

`from random import randint` - даний рядок підключає модуль генерації випадкових чисел

`minimum, maximum = 1, 10` - тут ми присвоїли значенням 1 та 10 імена `minimum` та `maximum` відповідно.

> Те ж саме можна було зробити за 2 окремих присвоєння
>
> `minimum = 1`
>
> `maximum = 10`

`random_number = randint(minimum, maximum)` - Python виконає аж 3 операції тут



1.   замінить `minimum` та `maximum` значеннями
2.   викличе генератор `randint(1, 10)` який поверне випадкове число від 1 до 10
3.   отримане від генератора значення отримає ім'я `random_number`
