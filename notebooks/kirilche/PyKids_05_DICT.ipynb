# dict

Dictionary (dict) - словник, або набір пар 'ключ : значення'

Словник - це складний тип даних, який


*   є послідовність (sequence)
*   змінний (mutable)
*   складається з окремих елементів (container type)
*   перший елемент пари (ключ) має бути унікальним

```python
my_dict = {"математика": 12, "інформатика": 11}
print(my_dict)
```

Щоб додати нове значення до словника, пишемо назву ключа в квадратних дужках, а значення після знака присвоїти.

```python
my_dict["історія"] = 10
print(my_dict)
```

Є 2 способи отримання значення назад із словника.


*   Звичайний (може викликати помилку)
*   Безпечний



```python
# print(my_dict["хімія"])
# Викличе помилку
# ----> 1 print(my_dict["хімія"])
# KeyError: 'хімія'
```

```python
# Безпечний спосіб, не викличе помилку, а поверне стандартне значення. Тут це '1'
print(my_dict.get("хімія", 1))
```

# update()
Словник не має методу **append()** тому що ключі - унікальні. Логіка тут наступна: update()

*   додасть ті пари "ключ:значення", ключів яких немає
*   перезапише значення  в тих парах "ключ:значення", ключі яких є

для прикладу візьмемо такий словник

```python
my_friends = {
    "Андрій": {"вік": 25, "професія": "фронтенд", "улюблене": ["JavaScript", "React"]},
    "Уляна": {"вік": 28, "професія": "дата інженер", "улюблене": ["Python", "NymPy", "SciPy"]}
}
```

Наступний код додасть одну пару та змінить одну існуючу. Оскільки ключ "Андрій" вже є в словнику, а ключа "Оксана" ще немає.

```python
new_friends = {
    "Андрій": {"вік": 40, "професія": "фронтенд", "улюблене": ["JavaScript", "TypeScript", "Python"]},
    "Оксана": {"вік": 20, "професія": "бекенд", "улюблене": ["Go", "SQL", "Python"]}
}
import json
print("ДО виконання update()")
print(json.dumps(my_friends, ensure_ascii=False, indent=2))
my_friends.update(new_friends)
print("ПІСЛЯ виконання update()")
print(json.dumps(my_friends, ensure_ascii=False, indent=2))
```

# del
видалення зі словнику відбувається так само як із списку

```python
del my_friends["Уляна"]
print(my_friends)
```

```python
# nesting

Вкладені структури даних адресуються послідовними [...][...][...]
```

```python
print(my_friends['Оксана'])
print(my_friends['Оксана']['улюблене'])
print(my_friends['Оксана']['улюблене'][0])
```

# keys()
Повертає всі ключі без значень

```python
print(my_friends.keys())
```

# values()
Повертає всі значення без ключів

```python
print(my_friends.values())
```

# items()

Найважливіший метод, без якого цикл зі словником не працюватиме (або незручний)

```python
for k, v in my_friends.items():
  print(f"Ключ '{k}' має значення {v}")
```
