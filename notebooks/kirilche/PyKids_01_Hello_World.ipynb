Додаємо змінні

```python
greeting = "Hello"
name = "World"
print(greeting, name)
```

Скористаємось спеціальним параметром функції, який зветься end та вказує на те, чим завершити друк. Зазвичай це спеціальний символ переходу на новий рядок - '\n'

```python
greeting = "Hello"
name = "World"
print(greeting, name, end="!")
```

А тепер змінимо змінні

```python
greeting = "Привіт"
name = "Світе"
print(greeting, name, end="!")
```

Додамо кому. Теж за допомогою спеціального параметру. Цього разу стандартне значення було ' ' (пробіл)

```python
greeting = "Привіт"
name = "Світе"
print(greeting, name, end="!", sep=", ")
```

Зверніть увагу, що якщо змінити порядок end та sep, то нічого по суті не зміниться.

```python
greeting = "Привіт"
name = "Світе"
print(greeting, name, sep=", ", end="!")
```

Але параметри з іменами завжди ідуть після анонімних. Виправте помилку, щоб піти далі.


> SyntaxError: positional argument follows keyword argument

Значить, що порядок аргументів (значень параметрів функції) - не коректний

Видаліть перший аргумент `sep=", "` або перенесіть його в кінець списку аргументів.



```python
greeting = "Привіт"
name = "Світе"
print(sep=", ", greeting, name, end="!")
```

В основному це все, але є ще одна версія, яка демонструє дещо іншу ідею. Ідею того, що рядки можна додавати. Як числа!

```python
greeting = "Привіт"
name = "Світе"
print(greeting + name)
```

Цього разу спеціальний параметр **sep** не працює, бо ми маємо один єдиний рядок (речення) після операції + (склеювання, або "конкатенація")

Тому знак коми та пробіл треба додати власноруч. Ну і заодно знак оклику в кінці нашого **речення**.

```python
greeting = "Привіт"
name = "Світе"
print(greeting + ", " + name + "!")
```

Або ж із нашим старим знайомим **end**

```python
greeting = "Привіт"
name = "Світе"
print(greeting + ", " + name, end="!")
```
